```python
# OPTION 2 with a loop

### CREATE A FUNCTION TO SOLVE THE ISSUE
# It is limited to 100 tracks, we will have to fix it:

def get_playlist_tracks(username, playlist_id):
    
    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']
    
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    
    return tracks

tracks = get_playlist_tracks("spotify", "2S2bAjLaE0CN2V0cTwAeEL")
```

```python
import pandas as pd
```

```python
# to solve the "limit 100" problem I'll create a empty list where I store all features from track_ids
# using a for loop every 100 track_ids and extend it in the empty list
# next step I  convert the audio_f list to a dataframe "df"

audio_f = []

for i in range(int(len(track_ids)/100)):
    x=0
    y=100
    audio_f.extend(sp.audio_features(track_ids[x:y]))
    x += 100
    y += 100

df = pd.DataFrame(audio_f)

df
```
